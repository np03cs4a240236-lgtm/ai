```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd
dataset = pd.read_csv("/content/drive/MyDrive/Concept and technologies of AI/Human_Development_Index_Dataset.csv")
unique_years = dataset['year'].unique()
print(unique_years)
hdi_2022_df = dataset[dataset['year'] == 2022]
hdi_2022_df.head(10)
```

```python
#Displaying the first 10 rows of the 2022 dataset.
first_rows=hdi_2022_df.head(10)
print(first_rows)
#Counting the number of rows and columns.
rows= hdi_2022_df.shape[0]
columns = hdi_2022_df.shape[1]
print("Number of rows:", rows)
print("Number of columns:", columns)
#Listing all column names and their data types.
dataset[dataset['year'] == 2022].dtypes
```

3. Missing Values & Data Cleaning:

```python
# Check missing values
print(hdi_2022_df.isna().sum())

# Check duplicates
print("Duplicates:", hdi_2022_df.duplicated().sum())

# Clean data
hdi_2022_df = hdi_2022_df.copy()
hdi_2022_df.replace(['–', '-', 'NA', ''], pd.NA, inplace=True)
hdi_2022_df = hdi_2022_df.drop_duplicates()

print("Cleaning done!")
```

```python
# Computing basic statistics for HDI
mean_hdi = hdi_2022_df['hdi'].mean()
median_hdi = hdi_2022_df['hdi'].median()
std_hdi = hdi_2022_df['hdi'].std()

#print("Mean HDI (2022):", mean_hdi)
#print("Median HDI (2022):", median_hdi)
#print("Standard Deviation of HDI (2022):", std_hdi)
#Identifying the country with the highest HDI in 2022.
highest_hdi_country = hdi_2022_df.loc[hdi_2022_df['hdi'].idxmax(), 'country']
#print("Country with highest HDI in 2022:", highest_hdi_country)
lowest_hdi_country = hdi_2022_df.loc[hdi_2022_df['hdi'].idxmin(), 'country']
#print("Country with lowest HDI in 2022:", lowest_hdi_country)
```

```python
high_hdi = hdi_2022_df[hdi_2022_df['hdi'] > 0.800]
print(high_hdi)
high_hdi_sorted = high_hdi.sort_values(by='gross_inc_percap',ascending=False)
print(high_hdi_sorted)
high_hdi_sorted.head(10)

```

6. Adding HDI Category Column:

```python

def classify_hdi(hdi):
    if hdi < 0.550:
        return "Low"
    elif 0.550 <= hdi <= 0.699:
        return "Medium"
    elif 0.700 <= hdi <= 0.799:
        return "High"
    else:
        return "Very High"

#creating new HDI category column
hdi_2022_df['hdi_category'] = hdi_2022_df['hdi'].apply(classify_hdi)
print(hdi_2022_df['hdi_category'].value_counts())
#checking rows
hdi_2022_df[['country', 'hdi', 'hdi_category']].head(10)
#saving final dataframe as csv
hdi_2022_df.to_csv("hdi_category_added.csv", index=False)
print("File saved as: hdi_category_added.csv")
```

2. Composite Development Score:

```python
# List of South Asian countries
south_asia_list = ["Afghanistan", "Bangladesh", "Bhutan", "India","Maldives", "Nepal", "Pakistan", "Sri Lanka"]

# Filter South Asian countries
south_asia_df = hdi_2020_2022[hdi_2020_2022['country'].isin(south_asia_list)].copy()
# Create Composite Score
south_asia_df.loc[:, 'Composite_Score'] = ( 0.30 * south_asia_df['life_expectancy'] +0.30 * south_asia_df['gross_inc_percap'])
# Use only 2022 data for ranking
south_asia_2022 = south_asia_df[south_asia_df['year'] == 2022]

# Sort by Composite Score
south_asia_ranked = south_asia_2022.sort_values(
    by='Composite_Score',
    ascending=False
)

# Display ranking
south_asia_ranked[['country', 'Composite_Score', 'hdi']]
# Select top 5 countries
top5 = south_asia_ranked.head(5)

plt.figure()
sns.barplot(
    data=top5,
    x='Composite_Score',
    y='country'
)
plt.xlabel("Composite Development Score")
plt.ylabel("Country")
plt.title("Top 5 South Asian Countries by Composite Development Score (2022)")
plt.show()
```

3. Outlier Detection:

```python
south_asia_2022 = south_asia_data[south_asia_data['year'] == 2022].copy()

# Calculate Q1, Q3, and IQR for HDI
Q1_hdi = south_asia_2022['hdi'].quantile(0.25)
Q3_hdi = south_asia_2022['hdi'].quantile(0.75)
IQR_hdi = Q3_hdi - Q1_hdi

hdi_lower = Q1_hdi - 1.5 * IQR_hdi
hdi_upper = Q3_hdi + 1.5 * IQR_hdi

# Calculate Q1, Q3, and IQR for GNI per Capita
Q1_gni = south_asia_2022['gross_inc_percap'].quantile(0.25)
Q3_gni = south_asia_2022['gross_inc_percap'].quantile(0.75)
IQR_gni = Q3_gni - Q1_gni

gni_lower = Q1_gni - 1.5 * IQR_gni
gni_upper = Q3_gni + 1.5 * IQR_gni

# Identify outliers
outliers = south_asia_2022[(south_asia_2022['hdi'] < hdi_lower) | (south_asia_2022['hdi'] > hdi_upper) |(south_asia_2022['gross_inc_percap'] < gni_lower) | (south_asia_2022['gross_inc_percap'] > gni_upper)]


# Create scatter plot with highlighted outliers
plt.figure(figsize=(10, 6))

# Normal points
normal = south_asia_2022[~south_asia_2022.index.isin(outliers.index)]
plt.scatter(normal['gross_inc_percap'], normal['hdi'], color='blue', s=80, alpha=0.6, label='Normal Countries')

# Outliers
plt.scatter(outliers['gross_inc_percap'], outliers['hdi'],  color='red', s=150, alpha=0.8, label='Outliers', edgecolors='black', linewidths=2)

plt.xlabel('GNI per Capita ($)')
plt.ylabel('HDI')
plt.title('Outlier Detection: GNI per Capita vs HDI (South Asia 2022)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

4. Exploring Metric Relationships:

```python
# Use South Asia data
south_asia_df = south_asia_data.copy()

# Remove missing values
south_asia_clean = south_asia_df[['hdi', 'gender_development', 'life_expectancy']].dropna()

# Calculate Pearson correlations
corr_gender = south_asia_clean['gender_development'].corr(south_asia_clean['hdi'])
corr_life = south_asia_clean['life_expectancy'].corr(south_asia_clean['hdi'])
# Create scatter plots with trendlines
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Gender Development vs HDI
sns.regplot(data=south_asia_clean, x='gender_development', y='hdi', ax=axes[0],
            scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
axes[0].set_title(f'Gender Development vs HDI\n(Correlation: {corr_gender:.4f})')
axes[0].set_xlabel('Gender Development Index')
axes[0].set_ylabel('HDI')

# Plot 2: Life Expectancy vs HDI
sns.regplot(data=south_asia_clean, x='life_expectancy', y='hdi', ax=axes[1],
            scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
axes[1].set_title(f'Life Expectancy vs HDI\n(Correlation: {corr_life:.4f})')
axes[1].set_xlabel('Life Expectancy (years)')
axes[1].set_ylabel('HDI')

plt.tight_layout()
plt.show()
```

5. Gap Analysis:

```python
dataset['GNI_HDI_Gap'] = dataset['gross_inc_percap'] - dataset['hdi']
south_asia = dataset[
    dataset['country'].isin([
        'Afghanistan', 'Bangladesh', 'Bhutan', 'India',
        'Maldives', 'Nepal', 'Pakistan', 'Sri Lanka'
    ])
]

south_asia_latest = south_asia.sort_values('year').groupby('country').tail(1)
south_asia_latest.sort_values('GNI_HDI_Gap', ascending=False)[
    ['country', 'GNI_HDI_Gap']
]
south_asia_latest.sort_values('GNI_HDI_Gap', ascending=True)[
    ['country', 'GNI_HDI_Gap']
]
top_positive = south_asia_latest.sort_values('GNI_HDI_Gap', ascending=False).head(3)
top_negative = south_asia_latest.sort_values('GNI_HDI_Gap', ascending=True).head(3)

gap_data = pd.concat([top_positive, top_negative])
plt.figure(figsize=(8,5))
sns.barplot(
    data=gap_data,
    x='GNI_HDI_Gap',
    y='country'
)
plt.title('Top 3 Positive and Negative GNI–HDI Gaps in South Asia')
plt.xlabel('GNI – HDI Gap')
plt.ylabel('Country')
plt.show()
```
